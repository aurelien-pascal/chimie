# Évolution temporelle d'un système chimique

##### Chargement des modules utiles

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

## Synthèse d'un éther-oxyde

```python
t = np.array([0,10,20,30,40,60,90,120,240])
n = np.array([0.0,1.7e-4, 2.8e-4, 3.7e-4, 4.4e-4, 5.5e-4, 6.4e-4, 7.0e-4, 8.3e-4])
c = n/10e-3
```

#### Tracé de la concentration en fonction du temps

```python
plt.plot(t,c,'b+',ms=15)
plt.grid(which='both')
plt.show()
```

#### Détermination des vitesses

```python
v = np.zeros(len(t))
for i in range(len(t)-1):
    v[i] = ( c[i+1] - c[i] ) / ( t[i+1] - t[i] )
v[-1] = 0
print("v = ",v)
```

### Méthode différentielle

```python
c0 = 1.0e-1
plt.plot( np.log( ( c0 - c[:-1] ) ) , np.log(v[:-1]) ,'b+' ,ms=15)
plt.grid()
plt.show()
```

#### Reglin

```python

lr = linregress( np.log( ( c0 - c[:-1] ) ) , np.log(v[:-1]) )
a = lr.slope
b = lr.intercept
r = lr.rvalue**2
print("a = ",a)
print("b = ",b)
print("r2 = ",r)

f = lambda x:a*x+b
plt.plot( np.log( ( c0 - c[:-1] ) ) , np.log(v[:-1]) ,'b+', ms=15)
plt.plot( np.log( ( c0 - c[:-1] ) ) , f(np.log( ( c0 - c[:-1] ) )), 'r-' )
plt.grid()
plt.show()
```

### Méthode intégrale

```python
plt.plot( t, 1/(c0 - c) ,'b+' ,ms=15)
plt.grid()
plt.show()
```

```python
lr = linregress( t, 1/(c0 - c) )
a = lr.slope
b = lr.intercept
r = lr.rvalue**2
print("a = ",a)
print("b = ",b)
print("r2 = ",r)

f = lambda x:a*x+b
plt.plot( t, 1/(c0 - c) ,'b+',ms=15)
plt.plot( t , f(t), 'r-' )
plt.grid()
plt.show()
```

# Une question de pression

```python
t = np.array([0,20,40,60])
p = np.array([320,414,465,497]) * 1.0e5/760
plt.plot(t,p,'b+',ms=15)
plt.grid()
plt.show()
```

```python
p0 = p[0]
plt.plot( t, 1/(2*p0-p), 'b+', ms=15 )
plt.grid()
plt.show()
```

##### RL

```python
lr = linregress( t, 1/(2*p0-p) )
a = lr.slope
b = lr.intercept
r = lr.rvalue**2
print("a = ",a)
print("b = ",b)
print("r2 = ",r)

f = lambda x:a*x+b
plt.plot( t, 1/(2*p0-p) ,'b+',ms=15)
plt.plot( t,f(t), 'r-' )
plt.grid()
plt.show()
```
